Ejercicio Job Demand

```python
df_US = df.copy()
```

```python
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
```

```python
df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')

# El resultado en la tabla no se puede ordenar (job_posted_month) con sort_values
# Entonces por este motivo se usa index metodos
```

```python
# Se guarda el DF en una variable y se usa reset_index para luego ordenar estos.
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_US_pivot.reset_index(inplace=True)
df_US_pivot['month #'] = pd.to_datetime(df_US_pivot['job_posted_month'], format=('%B')).dt.month
df_US_pivot['month #']
```

```python
df_US_pivot.head()  # Se agrega la columna al final de la tabla
```

```python
df_US_pivot.sort_values('month #', inplace=True) 
# Se ordena por la ultima columna y se guarda en si misma
```

```python
df_US_pivot.set_index('job_posted_month', inplace=True)  
# Se establece 'job_posted month' como el nuevo index una vez ordenado.
# Se eliminan las columnas antes de esta columna

df_US_pivot.drop(columns='month #', inplace=True) 
# Se elimina la columna creada para el proceso de ordenar

df_US_pivot
```

```python
df_US_pivot.plot(kind='line')
```
