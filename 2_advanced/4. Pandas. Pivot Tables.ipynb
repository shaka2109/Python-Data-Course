```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# COUNT OF JOBS TITLES
```

```python
pv_df = df.copy()
```

```python
pv_df.pivot_table(index='job_title_short', aggfunc='size')
# Se puede conseguir con un groupby(), los dos como serie
```

```python
pv_df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
# 'aggfunc' aplican a la columna en 'values' 
# Se puede conseguir con un groupby() pero como una serie
```

```python
pv_df.pivot_table(values='salary_year_avg',index='job_title_short',columns='job_country',aggfunc='median')
# No se puede conseguir con un groupby()
```

```python
pv_df.pivot_table(values='salary_year_avg',index='job_country',columns='job_title_short',aggfunc='median')
# Se pueden reordenar facilmente cambiando index y columns
```

Buscaremos realizar una grafica con Pivot Table

```python
top_countries = df['job_country'].value_counts().head(6).index
top_countries
# Se guardan los nombres de los paises en un array
```

```python
df_job_countries = df.pivot_table(
               values='salary_year_avg',
               index='job_country',
               columns='job_title_short',
               aggfunc='median')
df_job_countries = df_job_countries.loc[top_countries]
df_job_countries
# De la pivot table se filtran los top_countries con loc[]
```

```python
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']
df_job_countries = df_job_countries[job_titles]
df_job_countries
# Filtramos los tres principales trabajos creando una lista y llamandalos en el DF
```

```python
df_job_countries.plot(kind='bar')
# Graficamos con pandas
```

Ejercicios de Pivot Table

```python
# Promedio de salario por tipo de trabajo

df_tabla1 = df.pivot_table(values='salary_year_avg', index='job_title_short')
df_tabla1.sort_values(by='salary_year_avg',ascending=False)
```

```python
# Promedio de salario por país de la empresa

df_tabla2 = df.pivot_table(values='salary_year_avg', index='job_country')
df_tabla2.sort_values(by='salary_year_avg',ascending=False).head()
```

```python
# Promedio de salario por tipo de empleo y país

df_tabla3 = df.pivot_table(values='salary_year_avg',
                           index='job_country',
                           columns='job_schedule_type',
                           aggfunc='mean')
df_tabla3.head()
```

```python
# Reemplazar valores faltantes con 0 (agregar 'fill_value=)

df_tabla4 = df.pivot_table(values='salary_year_avg',
                           index='job_country',
                           columns='job_schedule_type',
                           aggfunc='mean',
                           fill_value = 0)
df_tabla4.head()
```

```python
# Mostrar salario máximo por país

df_tabla5 = df.pivot_table(values='salary_year_avg',
                           index='job_country',
                           columns='job_title_short',
                           aggfunc='max',
                           fill_value = 0)
df_tabla5.head()
```

```python
# Mostrar varias métricas a la vez

df_tabla6 = df.pivot_table(values='salary_year_avg',
                           index='job_country',
                           aggfunc=['max','min','mean'],
                           fill_value = 0)
df_tabla6.head()
```

```python
# Contar número de registros (empleos) por país y tipo de empleo

df_tabla7 = df.pivot_table(values='job_title_short',
                           index='job_country',
                           columns='job_schedule_type',
                           aggfunc='count',
                           fill_value=0)
df_tabla7.head()
```

```python
# Promedio de salario por pais y tipo de empleo, y redondear los valores del pivot.

df_tabla8 = df.pivot_table(values='salary_year_avg',
                           index='job_country',
                           columns='job_schedule_type',
                           aggfunc='mean',
                           fill_value=0) #Redondea los resultados más claro (sin decimales).
df_tabla8 = df_tabla8.round(0)
df_tabla8.head()
```

```python
# Create a pivot table to count the number of job postings for each job_title_short and job_country. 
# Display the first 10 rows of the resulting pivot table.

tabla9 = df.pivot_table( 
                         index=['job_title_short','job_country'],
                         aggfunc='size',
                        )
tabla9
```

```python
# Create a pivot table to find the maximum, minimum, and average salary_year_avg for each company_name. 
# Display the first 5 rows of the resulting pivot table.

tabla10 = df.pivot_table(values='salary_year_avg',
                         index='company_name',
                         aggfunc=['max','min','mean'],
                        )
tabla10 = tabla10.round(0)
tabla10.head()
```

```python
# Create a pivot table to calculate the median of salary_year_avg for each company_name and job_country:

# Filtrar el DF para solo 'Data Scientist'
tabla11 = df[df['job_title_short']=='Data Scientist'].copy()

# Creacion de tabla dinamica
tabla12 = tabla11.pivot_table(values='salary_year_avg',
                         index=['job_country','company_name'],
                         aggfunc='median'
                        )
tabla12 = tabla12.round(0)

# Filtrar tabla dinamica para salarios promedio mayores a 200.000
tabla_filtrada = tabla12[tabla12['salary_year_avg'] > 200000]
tabla_filtrada.head(10)
```
