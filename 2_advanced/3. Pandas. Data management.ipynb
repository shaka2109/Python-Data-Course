FUNCION COPY()

```python
# El metodo copy crea una copia nueva que puedes alterar sin que cambien datos del dataframe original
# cuando usas 'df_altered = df' y realizas operaciones en el primero, estas operaciones se aplican
# a df (dataframe original), entonces con .copy() se evita esto.

df_altered = df.copy()
```

FUNCION GROUPBY()

```python
nuevo_df = df.copy()
```

```python
# Calcular el salario promedio por tipo de trabajo
# Agrupa por 'job_title' y calcula la media de 'salary_year_avg'.

nuevo_df.groupby('job_title_short')['salary_year_avg'].mean().sort_values(ascending=False)
```

```python
# Calcular el salario máximo por país de la empresa.
# Agrupa por 'job_location' y obtén el salario máximo.

nuevo_df.groupby('job_location')['salary_year_avg'].max().sort_values(ascending=False)
```

```python
# Contar cuántos empleos hay por país de residencia
# Cuenta el número de registros por 'search_location'.

nuevo_df.groupby('search_location')['job_title'].size().sort_values(ascending=False)
```

```python
# Mostrar el salario promedio por tipo de empleo.
# Agrupa por 'job_schedule_type' y calcula el promedio de 'salary_year_avg'.

nuevo_df.groupby('job_schedule_type')['salary_year_avg'].mean().sort_values(ascending=False)
```

```python
# Mostrar varias estadísticas a la vez
# Agrupa por 'job_title' y calcula el salario promedio, mínimo y máximo usando el método .agg().

nuevo_df.groupby('job_title_short')['salary_year_avg'].agg(['min','max'])
```

```python
# Agrupar por dos columnas
# Agrupa por 'company_location' y 'employment_type', calcula el salario promedio para cada combinación.

nuevo_df.groupby(['job_location','job_schedule_type'])['salary_year_avg'].mean()
```

```python
# Convertir el resultado de groupby a DataFrame
# Convierte el resultado agrupado de 'job_title' en un nuevo DataFrame con columnas renombradas.

nuevo2_df = nuevo_df.groupby('job_title_short')['salary_year_avg'].mean().reset_index()
nuevo2_df.rename(columns = {'salary_year_avg':'avg_salary'}, inplace = True)
nuevo2_df.head(5)
# groupby() arroja una serie en una variable guardada.
# Al agregar .reset_index() lo convierte en DataFrame.
```

```python
# Contar cuántos valores únicos hay en cada grupo
# Agrupa por 'job_location' y cuenta cuántos títulos de trabajo diferentes hay por país.

nuevo_df.groupby('job_location')['job_title'].nunique()
```

```python
# Calcular el porcentaje de participación de cada país en el total de empleos
# Usa groupby() y count() y calcular número de empleos por país, luego divide entre el total general.

nuevo_df['job_location'].value_counts(normalize=True)*100
```
